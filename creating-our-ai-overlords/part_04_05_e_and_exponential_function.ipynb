## Plot `2^n`


```python
x = np.arange(0, 10, .1)

x
```

```python
y = []

base = 2 

for val in x:
    y.append(math.pow(base, val))
```

```python
plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y = 2^x')

plt.grid(axis = 'both')
plt.show()
```

## Derivative

$\huge\frac{2^{t+\Delta} - 2^t}{\Delta}$

<br><br>
$= \huge\frac{2^{t}2^{\Delta} - 2^t}{\Delta}$

<br><br>
$= \huge 2^t (\frac{2^{\Delta} - 1}{\Delta})$


<br><br>

Notice how the values in the parentheses don't depend at all on $\large t$.

We'll call the result that's in the parentheses the ***Base Proportionality Constant***


## Base Proportionality Constant



$\huge \frac{2^{\Delta} - 1}{\Delta}$

```python
def get_base_p_constant(base):
    delta = 0.00001
    return (math.pow(base, delta) - 1) / delta
```

```python
base = 2
get_base_p_constant(base)
```

## Derivative Code

$\huge 2^t (\frac{2^{\Delta} - 1}{\Delta})$

```python
def get_derivative(base, power):
    
    base_p_constant = get_base_p_constant(base)

    return math.pow(base, power) * base_p_constant
```

```python
get_derivative(2, 3)
```

## Plot `2^x` and its derivative

```python
derivatives = []

base = 2

for val in x:
    derivatives.append(
        get_derivative(base, val)
    )
```

```python
plt.plot(x, y, color='red', label='2^x')
plt.plot(x, derivatives, color='blue', label='Derivative of 2^x')

plt.legend(loc='upper center')

#plt.legend([y, derivatives], ['u','v'])

plt.xlabel('x')
plt.ylabel('y')

plt.grid(axis = 'both')
plt.show()
```
